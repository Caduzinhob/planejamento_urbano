# Visualiza√ß√£o de Estat√≠sticas do Grafo
Este notebook permite carregar uma inst√¢ncia `.ins`, calcular as estat√≠sticas do grafo e exibi-las em formato de tabela com `pandas`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from ler_arquivo import ler_arquivo_ins_arquivo
from calcular_estatisticas import calcular_estatisticas
from geracao_matriz import floyd_warshall
```

```python
arquivo = [
    r"D:\Meus arquivos\Documentos\GitHub\planejamento_urbano\ins\09.ins"
]
```

```python
caminho_exemplo = arquivo[0] 

grafo_exemplo = ler_arquivo_ins_arquivo(caminho_exemplo)
matriz_exemplo = floyd_warshall(grafo_exemplo)
estatisticas_exemplo = calcular_estatisticas(grafo_exemplo, matriz_exemplo)

df_exemplo = pd.DataFrame(estatisticas_exemplo.items(), columns=["M√©trica", "Valor"])
display(df_exemplo.style.set_caption(f"üìå Estat√≠sticas da Inst√¢ncia: {os.path.basename(caminho_exemplo)}").set_properties(**{'text-align': 'left'}))
```

```python
metricas_numericas = {k: v for k, v in estatisticas_exemplo.items() if isinstance(v, (int, float))}

plt.figure(figsize=(10, 6))
plt.bar(metricas_numericas.keys(), metricas_numericas.values(), color='skyblue')
plt.title("üìä Estat√≠sticas da Inst√¢ncia 02.ins", fontsize=14)
plt.ylabel("Valor")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
